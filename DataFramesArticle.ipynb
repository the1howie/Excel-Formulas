## Fruit Slicing Ninja

This code `fruit_sales.loc['Oranges']` does not work, unless you set the Fruit column as the index column:

```python
# fruit_sales.loc[['Oranges']]
```

```python
# Set column 'Fruit' as the index column
fruit_sales.set_index('Fruit', inplace=True)

display(fruit_sales)
```

```python
fruit_sales.loc[['Oranges']]
```

```python
fruit_sales.loc[['Oranges', 'Pears', 'Tomatoes']]
```

```python
fruit_sales.loc['Oranges':'Tomatoes']
```

### Slicing with integer indexing.

`iloc` is pure integer-location based indexing.

You can still work with the Fruit column as the index column, but this looks cleaner.


```python
# Reset index back to integer indexing.
fruit_sales.reset_index(inplace=True)

display(fruit_sales)
```

```python
fruit_sales.iloc[6:13, 1:3]  # ['Cherries':'Figs', ['Qtr1', 'Qtr2']]
```

```python
fruit_sales.iloc[:, 1]
```

```python
fruit_sales.iloc[:, [1, 4]]
```

```python
fruit_sales.iloc[:, 1:5]
```

```python
fruit_sales.iloc[2:5] #['Avocados':'Blackberries']
```

```python
fruit_sales.iloc[:3, [0, 1, 4]] #[:'Avocados', ['Qtr1', 'Qtr4']]
```

```python
fruit_sales.iloc[:, :]
```
